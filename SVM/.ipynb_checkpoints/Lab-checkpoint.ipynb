# Importation de package

```python
import numpy as np

#couleur map
from matplotlib.pyplot import subplots, cm

#sélection de modèle
import sklearn.model_selection as skm

from ISLP import load_data, confusion_table
```

```python
# Support Vector Classification dans la module SVM
from sklearn.svm import SVC

from ISLP.svm import plot as plot_svm

from sklearn.metrics import RocCurveDisplay

#Fonction de ROC
roc_curve = RocCurveDisplay.from_estimator
```

# 1.Support Vector Classifier

```python
#Générateur de graine pour assurer la reproductibilité
rng = np.random.default_rng(1)

# Matrice dont les nombres suivent la loi gaussien
X = rng.standard_normal((50,2))

# Les classes
y = np.array([-1]*25 + [1]*25)
X[y==1] += 1
```

```python
#Voir la forme des données
fig, ax = subplots(figsize = (8,8))
ax.scatter (X[:,0],
            X[:,1],
            c = y,
            cmap = cm.coolwarm)
```

Les données en rouge se situent à droite et en haut tandis que les points en bleu sont plus au centre. 

```python
#Application du SVM linéaire
svm_linear_1 = SVC(C=10, kernel='linear')
svm_linear_1.fit(X,y)
```

```python
fig, ax = subplots(figsize=(8,8))
plot_svm(X,
         y,
         svm_linear_1,
         ax = ax)
```

Dans le contexte des machines à vecteurs de support (SVM), les "support vectors" (vecteurs de support) sont les données qui se trouvent le plus près de l'hyperplan de décision. Ce sont les points de données qui "soutiennent" l'hyperplan. 

Ce sont les échantillons les plus difficiles à classer et ils définissent la marge de séparation entre les classes.

Et si on utilise une valeur très faible de C?

```python
svm_linear_2 = SVC(C=0.001, kernel='linear')
svm_linear_2.fit(X,y)
fig, ax = subplots(figsize=(8,8))
plot_svm(X,
         y,
         svm_linear_2,
         ax = ax)
```

Le fait de diminuer la marge va augmenter les supports de vecteurs. Puisque la marge est devenue très grande.

# Montrons les coefficients Beta  : 


```python
#Coefficients
svm_linear_1.coef_
```

```python
# Validation croiséé en 5 sous-ensembles
kfold = skm.KFold (5,
                   random_state=0,
                   shuffle=True)#données divisée avant la division en plis)
```

```python
# On retourne le svm_linéaire dans une grille de validation de croiséé
grid = skm.GridSearchCV(svm_linear_1,
                       {'C':[0.001,0.01,0.1,1,5,10,100]},
                        refit=True #modèle final ajusté automatiquement par la meilleure valeure de C,
                        cv=kfold,
                        scoring='accuracy')
grid.fit(X,y)
grid.best_params_
```

```python
# Scores moyens de validation croisée pour les hyperparamètres
grid.cv_results_[('mean_test_score')]
```

La quatrième score correspond à la valeur de C=1. Elle permet de maximiser la performance de C tout en le minimisant(moins de contrainte).  

# Voir le meilleur estimateur sur les données test

```python
X_test = rng.standard_normal((20,2))
y_test = np.array([-1]*10 + [1]*10)
X_test[y_test==1] += 1 #données négatives restent inchangées
```

```python
best_ = grid.best_estimator_
best_
```

```python
# Matrice de confusion
y_test_hat = best_.predict(X_test)
confusion_table(y_test_hat, y_test)
```

Les "8" correspondent aux vrais négatifs (VN) : le modèle a correctement prédit 8 instances de la classe -1.
Les "6" correspondent aux vrais positifs (VP) : le modèle a correctement prédit 6 instances de la classe 1.
Les "2" correspondent aux faux positifs (FP) : le modèle a incorrectement prédit 2 instances de la classe -1 comme appartenant à la classe 1.
Les "4" correspondent aux faux négatifs (FN) : le modèle a incorrectement prédit 4 instances de la classe 1 comme appartenant à la classe -1.

70% de la classification a été la bonne

# qu'est-ce qui se passe si on utilise C=0.001

```python
svm = SVC(C=0.001,
         kernel='linear').fit(X, y)
y_test_hat = svm.predict(X_test)
confusion_table(y_test_hat, y_test)
```

Dans ce cas, il y a 60% de bon classement. Ce qui justifie que la C=1 est la bonne valeur.

# séparation de données SVM

```python
X[y==1] += 2 ; 
fig, ax = subplots(figsize=(8,8))
ax.scatter(X[:,0], X[:,1], c=y, cmap=cm.coolwarm) ; 
```

La grande séparation est maintenant tangible.

```python
svm = SVC(C=1e5, kernel='linear').fit(X,y)
y_hat = svm.predict(X)
confusion_table(y_hat, y)
```

```python
fig, ax = subplots(figsize=(8,8))
plot_svm(X,
        y,
        svm,
        ax=ax)
```

Comme on peut voir, il y a uniquement trois supports de vecteurs et une bonne séparation de données. Mais celà peut entrainer un surajustement. On devrait essayer donc la méthode sur les données test.

```python
svm = SVC (C=(0.1), kernel = 'linear').fit(X,y)
y_hat = svm.predict(X)
confusion_table(y_hat, y)
```

```python
fig, ax = subplots(figsize=(8,8))
plot_svm(X,
y,
svm,
ax=ax)
```

La valeur faible de C n'entraine pas un changement sur la prédiction. Par contre, cela affecte le nombre de supports vecteurs.

# Conclusion :

**Vecteurs de Support** : Points déterminant la marge de séparation entre classes.

**Marge :** Écart entre les vecteurs de support et l'hyperplan, influence la généralisation.

**Paramètre C :** Contrôle le compromis entre marge large et erreur de classification faible. C petit, SVM maximise la marge même en tolérant l'erreur de classification. On peut avoir un modèle plus général mais tachée d'erreur. Et vice-versa

**Ajustement de C :** Un C optimal équilibre nombre de vecteurs de support et performance de classification

# 2. Support Vector Machine

```python
X = rng.standard_normal ((200, 2))
X[:100] += 2
X[100:150] -= 2
y = np.array([1]*150 + [2]*50)
```

```python
fig, ax = subplots(figsize=(8,8))
ax.scatter(X[:,0],
X[:,1],
c=y,
cmap=cm.coolwarm)
```

```python
#Radial kernel avec gamma = 1
(X_train,
X_test,
y_train,
y_test) = skm.train_test_split(X,
                              y,
                              test_size=0.5,
                              random_state=0)
svm_rbf = SVC(kernel="rbf", gamma = 1, C = 1)
svm_rbf.fit(X_train, y_train)
```

```python
fig, ax = subplots(figsize=(8,8))
plot_svm(X_train,
y_train,
svm_rbf,
ax=ax)
```

On peut voir qu'il y a une erreur de classification de l'étiquette verte. Lorsqu'on augmente C, ce problème peut être résolu mais au risque d'une frontière de décision irrégulière au risque d'un surajustement.

```python
svm_rbf = SVC(kernel="rbf", gamma=1, C=1e5)
svm_rbf.fit(X_train, y_train)
fig, ax = subplots(figsize=(8,8))
plot_svm(X_train,
y_train,
svm_rbf,
ax=ax)
```

```python
# Choix des paramètres avec VC
kfold = skm.KFold(5,
random_state=0,
shuffle=True)
grid = skm.GridSearchCV(svm_rbf,
{'C':[0.1,1,10,100,1000],
'gamma':[0.5,1,2,3,4]},
refit=True,
cv=kfold,
scoring='accuracy');
grid.fit(X_train, y_train)
grid.best_params_
```

```python
best_svm = grid.best_estimator_
fig, ax = subplots(figsize=(8,8))
plot_svm(X_train,
         y_train,
best_svm,
ax=ax)
y_hat_test = best_svm.predict(X_test)
confusion_table(y_hat_test, y_test)
```

7% sont mal classés par le SVM rbf.

# ROC Curves

SVM classe en fonction de la signe de la fonction de décision. Si c'est supérieur à 0, c'est classe positive. Sinon, dans les autres. C'est une classification one class VS others. 

```python
fig, ax = subplots(figsize=(8,8))
roc_curve(best_svm,
X_train,
y_train,
name='Training',
color='r',
ax=ax);
```

```python
svm_flex = SVC(kernel="rbf",
gamma=50,
C=1)
svm_flex.fit(X_train, y_train)
fig, ax = subplots(figsize=(8,8))
roc_curve(svm_flex,
X_train,
y_train,
name='Training $\gamma=50$',
color='r',
ax=ax);
```

traçage de la courbe de ROC sur l'échantillon test

```python
fig, ax = subplots(figsize=(8,8))
roc_curve(svm_flex,
    X_test,
    y_test,
    name='Test $\gamma=50$',
    color='b',
    ax=ax)
    fig;
```

```python
fig, ax = subplots(figsize=(8,8))
for (X_, y_, c, name) in zip(
    (X_train, X_test),
    (y_train, y_test),
    ('r', 'b'),
    ('CV tuned on training',
    'CV tuned on test')):
    roc_curve(best_svm,
        X_,
        y_,
        name=name,
        ax=ax,
        color=c)
```

Le surajustement n'est pas au rendez-vous

# SVM multi-classes

```python
X = rng.standard_normal((50,2))

# Les classes
y = np.array([-1]*25 + [1]*25)
```

```python
rng = np.random.default_rng(123)
random_data = rng.standard_normal((50,2))
random_data
```

```python
X = np.concatenate([X, random_data], axis=0)
```

```python
y = np.append(y, [0]*50)
```

```python
X[y==0, 1] += 2
```

```python
fig, ax = subplots(figsize=(8,8))
ax.scatter(X[:,0], X[:,1], c=y, cmap=cm.coolwarm);
```

Nous avons maintenant trois classes

```python
svm_rbf_3 = SVC(kernel="rbf",
C=10,
gamma=1,
decision_function_shape='ovo');
svm_rbf_3.fit(X, y)
fig, ax = subplots(figsize=(8,8))
plot_svm(X,
y,
svm_rbf_3,
scatter_cmap=cm.tab10,
ax=ax)
```

# Conclusion : 
Nombre d'hyperplan = nombre de classe - 1

# Application sur GENE

```python
Khan = load_data('Khan')
Khan['xtrain'].shape, Khan['xtest'].shape
```

C'est une donnée à haute dimension avec une faible observation.
Nous utiliserons une approche de vecteur de support pour prédire le sous-type de cancer en utilisant les mesures d'expression génique. Dans ce jeu de données, il y a un très grand nombre de caractéristiques par rapport au nombre d'observations. Cela suggère que nous devrions utiliser un noyau linéaire, car la flexibilité supplémentaire résultant de l'utilisation d'un noyau polynomial ou radial est inutile.

```python
khan_linear = SVC(kernel='linear', C=10)
khan_linear.fit(Khan['xtrain'], Khan['ytrain'])
confusion_table(khan_linear.predict(Khan['xtrain']),
Khan['ytrain'])
```

Nous constatons qu'il n'y a pas d'erreurs d'entraînement. En fait, cela n'est pas surprenant, car le grand nombre de variables par rapport au nombre d'observations implique qu'il est facile de trouver des hyperplans qui séparent complètement les classe

```python
confusion_table(khan_linear.predict(Khan['xtest']),
Khan['ytest'])
```

```python
il y a deux ereurs uniquement.
```
